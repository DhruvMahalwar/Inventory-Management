```python
# Importing necessary modules
import json 
import time

# Reading data from the "Records.json" file
fd = open("Records.json", 'r')
js = fd.read()
fd.close()

# Loading JSON data into a dictionary
records = json.loads(js)

# Displaying menu
print("------------------------Menu-----------------------")
for i in records.keys():
    print(i, ":", records[i]["Name"], "|", records[i]["Price"], "|", records[i]["Qn"])
print("---------------------------------------------------")

# Taking user input for the purchase
ui_name = str(input("Enter your name     : "))
ui_mail = str(input("Enter Mail ID       : "))
ui_ph = str(input("Enter Phone number  : "))
ui_pr = str(input("Enter Product ID    : "))
ui_qn = int(input("Enter Quantity     : "))

# Checking if the requested quantity is available
if ui_qn <= records[ui_pr]["Qn"]:
    print("---------------------------------------------------")
    print("                       BILL                        ")
    print("")
    print("Name                 : ", records[ui_pr]["Name"])
    print("Price                : ", "Rs", records[ui_pr]["Price"], )
    print("Quantity             : ", ui_qn)
    print("---------------------------------------------------")
    print("Billing Amount       : ", "Rs", records[ui_pr]["Price"] * ui_qn)
    print("---------------------------------------------------")

    # Updating sales information and reducing inventory quantity
    sales = ui_name + "," + ui_mail + "," + ui_ph + "," + ui_pr + "," + records[ui_pr]["Name"] + "," + str(
        ui_qn) + "," + str(ui_qn * records[ui_pr]["Price"]) + "," + time.ctime() + "\n"
    records[ui_pr]["Qn"] -= ui_qn
    
    # Writing sales information to "Sales.txt"
    fd = open("Sales.txt", 'a')
    fd.write(sales)
    fd.close()

else:
    # If requested quantity is not available, displaying a message
    print("Sorry we are not having enough Quantity in our Inventory")
    print("We are only having", records[ui_pr]["Qn"], "quantity.")
    print("---------------------------------------------------")
    want = input("Press Y/y to Purchase : ")

    if want == 'Y' or want == 'y':
        # If user still wants to purchase available quantity, proceed with the purchase
        print("---------------------------------------------------")
        print("                       BILL                        ")
        print("")
        print("Name                 : ", records[ui_pr]["Name"])
        print("Price                : ", "Rs", records[ui_pr]["Price"], )
        print("Quantity             : ", records[ui_pr]["Qn"])
        print("---------------------------------------------------")
        print("Billing Amount       : ", "Rs", records[ui_pr]["Price"] * records[ui_pr]["Qn"])
        print("---------------------------------------------------")

        # Updating sales information and setting inventory quantity to 0
        sales = ui_name + "," + ui_mail + "," + ui_ph + "," + ui_pr + "," + records[ui_pr]["Name"] + "," + str(
            records[ui_pr]["Qn"]) + "," + str(records[ui_pr]["Qn"] * records[ui_pr]["Price"]) + "," + time.ctime() + "\n"
        records[ui_pr]["Qn"] = 0

        # Writing sales information to "Sales.txt"
        fd = open("Sales.txt", 'a')
        fd.write(sales)
        fd.close()

    else:
        # If user does not want to proceed with the purchase, display a message
        print("Thanks!")

# Updating the inventory data in "Records.json"
js = json.dumps(records)
fd = open("Records.json", 'w')
fd.write(js)
fd.close()

# Displaying a thank you message
print("")
print("---------------------------------------------------")
print("     Thanks for your order, Inventory Updated!     ")
print("---------------------------------------------------")
```
